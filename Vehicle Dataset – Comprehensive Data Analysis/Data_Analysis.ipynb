In this column some values are simmiler so that i will combine that values: 


reguler unleaded ------> petrol

premium unleaded (reqired) ------> petrol

reguler unleaded (recomended) ------> petrol

flex-fuel (unleaded/E85) ------> flex-fuel

flex-fuel (premium unleaded required/E85) ------> flex-fuel

flex-fuel (premium unleaded recommended/E85) ------> flex-fuel

diesel ------> diesel

electric ------> electic

netural gas ------> gas


--> Reduces noise in the data and helps with faster and more stable training.

```python
change_name = {
    'regular unleaded': 'petrol',
    'premium unleaded (required)': 'petrol',
    'premium unleaded (recommended)': 'petrol',
    'flex-fuel (unleaded/E85)': 'flex-fuel',
    'flex-fuel (premium unleaded required/E85)': 'flex-fuel',
    'flex-fuel (premium unleaded recommended/E85)': 'flex-fuel',
    'diesel': 'diesel',
    'electric': 'electric',
    'natural gas': 'gas'
}

df['Engine Fuel Type'] = df['Engine Fuel Type'].map(change_name)
```

```python
df['Engine Fuel Type'].value_counts()
```

```python
df['Engine Cylinders'].value_counts()
```

In this column the 0.0 values are incorrect that i will remove so that data is more accurate.

```python
df = df[df['Engine Cylinders'] != 0.0]
```

```python
df['Engine Cylinders'].value_counts()
```

```python
df['Transmission Type'].value_counts()
```

In this column some values are unfrequent so i will manage them :


--> firestly i will convert values into lower case.
--> then combine unfrequent values that is 'automated_manual' and 'direct_drive' and assign new name 'other'.


---->Prevents overfitting from rare categories in ML and reduces dimensionality in encoding (like one-hot).

```python
df['Transmission Type'] = df['Transmission Type'].str.lower().str.strip()
```

```python
df['Transmission Type'] = df['Transmission Type'].replace({
    'direct_drive': 'other',
    'automated_manual': 'other' 
})
```

```python
df['Transmission Type'].value_counts()
```

```python
df['Driven_Wheels'].value_counts()
```

Same thing in this column also because 'all wheel drive' and four wheel drive are same thing that's
why i will combine them.


To reduce the number of categories (helpful for machine learning) and avoid confusion between similar drive types.

```python
df['Driven_Wheels'] = df['Driven_Wheels'].str.lower().str.strip()
```

```python
df['Driven_Wheels'] = df['Driven_Wheels'].replace({
    'four wheel drive': 'all wheel drive'
})
```

```python
df['Number of Doors'].value_counts()
```

In this column the 3.0 values are incorrect that why i will remove them.

```python
combo_counts = df['Market Category'].value_counts()
common_combos = combo_counts[combo_counts > 50].index

df['Market Category'] = df['Market Category'].apply(
    lambda x: x if x in common_combos else 'Other'
)
```

## DATA VISUALIZATION & ANALYSIS

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
top_makes = df['Make'].value_counts().head(10)
plt.figure(figsize=(12,6))
sns.barplot(x=top_makes.index, y=top_makes.values)
plt.title('Top 10 Vehicle Makes')
plt.ylabel('Count')
plt.xlabel('Make')
plt.show()
```

### Insights :

- Chevrolet is the most common vehicle make, with a significantly higher count than all others.

- Mazda, Cadillac, and GMC are the least common among the top 10, showing similar and lower frequencies.

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Engine HP'], kde=True, bins=30, color='skyblue')
plt.title('Distribution of Engine Horsepower')
plt.xlabel('Engine HP')
plt.ylabel('Frequency')
plt.show()
```

### Insights :

- Most vehicles have engine horsepower between 150 and 300, that is common and most frequent.

- Very few vehicles have horsepower above 600, showing a right-skewed distribution with rare high-power engines.

```python
plt.figure(figsize=(18,6))
style_hp = df.groupby('Vehicle Style')['Engine HP'].mean().sort_values(ascending=False)
sns.barplot(x=style_hp.index, y=style_hp.values)
plt.title('Average Engine HP by Vehicle Style')
plt.ylabel('Average Engine HP')
plt.xlabel('Vehicle Style')
plt.xticks(rotation=45)
plt.show()
```

### Insights :

- Coupes and Convertibles have the highest average engine horsepower, suggesting a focus on performance.

- Hatchbacks and Convertible SUVs have the lowest average horsepower, indicating they are more suited for daily or utility use.

- 4 Door SUVs and Sedans are best velue for money choices car styles.

```python
sns.countplot(data=df, x='Transmission Type')
plt.title('Transmission Type Distribution')
plt.ylabel('Count')
plt.xlabel('Transmission Type')
plt.show()
```

### Insights :

- Automatic transmissions are the most common, making up the majority of vehicles.

- May be Automatic transmission cars are chiper then manual or other on the other hand the company shifts the car to manual to automatic transmission. 

- Manual and other types are far less frequent, showing a strong preference for automatic systems.

```python
fuel_counts = df['Engine Fuel Type'].value_counts()
plt.pie(fuel_counts, labels=fuel_counts.index, autopct='%1.1f%%')
plt.title('Fuel Type Distribution')
plt.show()
```

### Insights : 

- Petrol is the dominant fuel type, used by over 90% of vehicles.

- Diesel, gas, and flex-fuel types are used very rarely, indicating limited fuel diversity.

```python
sns.scatterplot(data=df, x='Engine HP', y='MSRP', hue='Vehicle Size')
plt.yscale('log')
plt.title("MSRP vs Engine HP")
plt.show()
```

### Insights : 

- Higher engine horsepower generally leads to higher MSRP, showing a positive correlation.

- All vehicle sizes (compact, midsize, large) follow this trend, but large vehicles often cluster at higher horsepower and price ranges.

- Compact vehicles are spread across a wide range of engine horsepower, but their MSRP tends to stay lower than large vehicles.

- There is a clear gap in pricing, suggesting two distinct pricing tierspossibly separating economy and luxury/performance vehicle segments.

```python
top_models = df['Model'].value_counts().head(10)
plt.figure(figsize=(12,6))
sns.barplot(x=top_models.index, y=top_models)
plt.title('Top 10 Vehicle Models')
plt.xlabel('Model')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

### Insights : 

- Silverado 1500 is the most popular vehicle model, leading the top 10 in count.

- Pickup trucks like Tundra, F-150, and Sierra 1500 dominate the top spots, showing high demand for utility vehicles.

- Sedans like Accord and Civic also appear in the top 10, indicating a balanced mix of utility and family vehicles.

- Corvette, a performance car, has the lowest count in the top 10 but still maintains a strong presence.

```python
wheels = df['Driven_Wheels'].value_counts()
plt.pie(wheels, labels=wheels.index, autopct='%1.1f%%')
plt.title('Distribution of Driven Wheels')
plt.show()
```

### Insights : 

- Front-wheel drive is the most common, used in nearly 40% of vehicles.

- All-wheel drive comes next at 33.2%, indicating strong traction is important for many users.

- Rear-wheel drive is the least common, making up 27.2% of the distribution.

```python
market = df['Market Category'].value_counts().head(10)
plt.figure(figsize=(10,6))
sns.barplot(y=market.index, x=market.values)
plt.title('Top 10 Market Categories')
plt.xlabel('Count')
plt.ylabel('Market Category')
plt.show()
```

### Insights : 
 
- Crossover dominates the market categories by a large margin, showing its high consumer preference.

- Luxury and Flex Fuel are also popular, but their counts are far below Crossovers.

- Performance and High-Performance segments appear less frequently, indicating they are niche categories.

- Combined categories like Luxury, Performance and Exotic, High-Performance have lower representation, suggesting they target specific, premium audiences.

```python
fuel_style = pd.crosstab(df['Vehicle Style'], df['Engine Fuel Type'])
fuel_style.plot(kind='bar', stacked=True, figsize=(12,7))
plt.title('Fuel Type Distribution by Vehicle Style')
plt.xlabel('Vehicle Style')
plt.ylabel('Count')
plt.legend(title='Fuel Type')
plt.show()
```

### Insights : 

- Petrol is the most dominant fuel type across all vehicle styles, especially in Sedans and 4-door SUVs.

- Flex-fuel is most commonly found in pickup trucks like Crew Cab and Extended Cab models.

- Diesel usage is minimal overall but slightly higher in Wagons and Sedans.

- Gas fuel is almost negligible, with a tiny presence in only a few vehicle styles.

```python
plt.figure(figsize=(8,6))
sns.countplot(data=df, x='Vehicle Size')
plt.title('Distribution of Vehicle Sizes')
plt.xlabel('Vehicle Size')
plt.ylabel('Count')
plt.show()
```

### Insights : 

- Compact vehicles are the most common, followed closely by Midsize vehicles.

- Large vehicles are significantly less common compared to Compact and Midsize types.

- The distribution shows a clear preference for smaller vehicle sizes, suggesting consumer or market trends favoring compactness and efficiency.

```python

sns.boxplot(data=df, x='Transmission Type', y='Engine HP')
plt.title('Engine HP by Transmission Type')
plt.show()
```

### Insights :

- Manual and automatic transmissions have similar median engine horsepower, but automatics have slightly higher values.

- Vehicles labeled as "other" transmission types tend to have a wider range and higher maximum horsepower, including the highest outliers.

- Manual transmissions show the most outliers, indicating more variation in horsepower among those vehicles.

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Engine Cylinders', y='Engine HP', data=df)
plt.title('Engine HP by Number of Cylinders')
plt.xlabel('Engine Cylinders')
plt.ylabel('Engine Horsepower')
plt.show()
```

### Insights : 

- Engine horsepower generally increases with the number of cylinders, indicating a strong positive correlation.

- 6-cylinder and 8-cylinder engines have the widest variability in horsepower, with many outliers, suggesting a range of performance levels.

- 10, 12, and 16-cylinder engines have consistently high horsepower, with fewer extreme outliers, reflecting their performance-oriented nature.

- 3 and 4-cylinder engines are associated with the lowest horsepower and are likely used in economy or compact vehicles.

```python
yearly_counts = df['Year'].value_counts().sort_index()
plt.plot(yearly_counts.index, yearly_counts.values, marker='o')
plt.title('Number of Vehicles Over Years')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

### Insights :

- From 1990 to around 2013, the number of vehicles increased gradually with small fluctuations.

- A sharp spike occurred between 2014 and 2015, indicating a sudden surge in vehicle records or production.

- After peaking in 20152016, the count slightly declined in 2017, suggesting a potential drop in production, sales, or data recording.
